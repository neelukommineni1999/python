```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("EventsPractice").getOrCreate()

# Sample data
events_data = [
    (1, "Login", "USA"),
    (2, "Purchase", "India"),
    (3, "Logout", "USA"),
    (4, "Login", "Canada"),
    (5, "Purchase", "India")
]

# Column names
columns = ["event_id", "event_type", "country"]

# Create DataFrame
df = spark.createDataFrame(events_data, columns)

# Show first 5 rows
df.show(5)
```
