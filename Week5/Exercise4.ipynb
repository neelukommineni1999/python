```python
3.Explain why calling show() twice without caching may cause Spark to read and process the data twice.

If we use show() two times, Spark may process the data two times because each show() is a separate action. 
Without caching, Spark runs the whole pipeline again.
```

```python
4.Explain lazy evaluation using a cooking analogy.

Lazy evaluation is like writing a recipe before cooking.

Writing the recipe = transformations
Cooking the food = action

Spark waits until an action like show() is called before doing the actual work.
```
