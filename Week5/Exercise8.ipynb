```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *

# Create Spark session
spark = SparkSession.builder.appName("SchemaPractice").getOrCreate()

# Define schema
schema = StructType([
    StructField('event_id', IntegerType(), True),
    StructField('user_id', StringType(), True),
    StructField('event_type', StringType(), True),
    StructField('device', StringType(), True),
    StructField('country', StringType(), True),
    StructField('event_date', StringType(), True),
    StructField('session_time_sec', IntegerType(), True)
])

# Read CSV using schema
df = spark.read.csv(
    'events.csv',
    header=True,
    schema=schema
)

# Print schema
df.printSchema()
```

```python
from pyspark.sql import functions as F

null_count = df.filter(
    F.col('user_id').isNull()
).count()

print("NULL user_id rows:", null_count)
```

```python
df_clean = df.fillna({
    'user_id': 'anonymous'
})
```

```python
remaining_nulls = df_clean.filter(
    F.col('user_id').isNull()
).count()

print("Remaining NULL user_id rows:", remaining_nulls)
```

```python
inferSchema=True is risky because Spark guesses datatypes from the data, 
and if the source system changes a column format, 
the inferred datatype may change and break the pipeline.
```
