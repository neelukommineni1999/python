```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("EventsPractice").getOrCreate()

# Sample events data
events_data = [
    (1, "Login", "USA"),
    (2, "Purchase", "India"),
    (3, "Logout", "USA"),
    (4, "Login", "Canada"),
    (5, "Purchase", "India")
]

# Column names
columns = ["event_id", "event_type", "country"]

# Create DataFrame
df = spark.createDataFrame(events_data, columns)

# Show first 5 rows
df.show(5)

# Register DataFrame as temp view
df.createOrReplaceTempView("events")

# SQL query to count events by country
result = spark.sql("""
SELECT country, COUNT(*) AS event_count
FROM events
GROUP BY country
""")

# Show SQL result
result.show()
```
